```python
# Creating Level 1 visualizations again
# 1. Bar Chart for Today's Mood and Energy Levels in community
plt.figure(figsize=(8, 5))
today_bar_data = today_data[['Mood Score', 'User Score']].mean()
today_bar_data.plot(kind='bar', color=['skyblue', 'orange'], alpha=0.7)
plt.title("Today's Mood and Energy Levels")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.show()
```

```python
user1_today = today_data[today_data["user_id"]=="user_1"]
user1_today
```

```python
# Calculate the community average score
community_avg_score = today_data["User Score"].mean()
user1_today = today_data[today_data["user_id"]=="user_1"]
user1_today_score= user1_today["User Score"].mean()

# Generate a bar chart comparing individual and community average scores
plt.figure(figsize=(8, 5))
plt.bar(["Community Average", "your score"], [community_avg_score, user1_today["User Score"].mean()], 
        color=["blue", "orange"], alpha=0.7)
plt.title("Comparison of User's Average Score vs. Community Average")
plt.ylabel("Score")
plt.ylim(0, 100)
plt.show()

# Provide insights
print(f"The community average score for today's check-ins is {community_avg_score:.2f}.")
print(f"Your today's status score is {user1_today_score}.")
if user1_today["User Score"].mean() >= 70:
    print("Great job! You are performing well today.")
else:
    print("Let's strive to improve the overall scores tomorrow!")
```

```python
# encouragment and Recommodation Widget
top_aspect = "Mood" if today_bar_data['Mood Score'] > today_bar_data['User Score'] else "Energy"
encouragement = f"Great work! Your {top_aspect.lower()} score is excellent today! Keep it up!"
recommendation = (
    "Consider taking a few moments to stretch and reflect for better focus." 
    if top_aspect == "Energy" 
    else "A short walk can lift your spirits and boost productivity."
)

# Print Encouragement and Recommendation
print(encouragement)
print("Daily Recommendation:", recommendation)
```

```python
# Suggestion for Improvement
low_aspect = "Mood" if today_bar_data['Mood Score'] < today_bar_data['User Score'] else "Energy"
suggestion = f"Try focusing on activities that improve your {low_aspect.lower()} level today."
print(f"Improvement Suggestion:: {suggestion}")
```

```python
## weekly mood trends

# Group check-in data by ISO week and calculate the average Mood Intensity
weekly_mood = user_score.groupby(user_score['date'].dt.isocalendar().week)['Mood Score'].mean()

# Plot weekly mood trends
plt.figure(figsize=(10, 6))
weekly_mood.plot(kind='line', marker='o', title="Weekly Mood Trends")
plt.xlabel("Week Number")
plt.ylabel("Average Mood Intensity")
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Correlate cycle days with mood
cycle_days = np.arange(1, 8)  # Example cycle days
moods = ['Happy', 'Neutral', 'Irritable']
heatmap_data = np.random.randint(1, 5, size=(len(moods), len(cycle_days)))

sns.heatmap(heatmap_data, annot=True, xticklabels=cycle_days, yticklabels=moods, cmap="coolwarm")
plt.title("Mood by Cycle Days")
plt.xlabel("Cycle Days")
plt.ylabel("Mood")
plt.show()

# Meaning: This heatmap reveals patterns between cycle days and moods, enabling insights into how phases impact emotions.
```
